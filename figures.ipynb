```python
import matplotlib.pyplot as plt
import scipy.stats
import numpy as np
```

```python
distances_distribution = scipy.stats.norm(0.5, 0.05)
memorizations_distribution = scipy.stats.norm(1, 0.01)
```

```python
xvalues = np.linspace(0, 1, 1000)
yvalues = distances_distribution.pdf(xvalues) + memorizations_distribution.pdf(xvalues) / 10
```

```python
ninetysixth = 0.972
vertical = 3.9
ninetyfifth = 0.6
textsize = 20
linewidth=2.0
```

```python
fig = plt.figure(figsize=(16, 3))
plt.plot(xvalues, yvalues)
plt.fill_between(xvalues, yvalues, color='lightblue')
plt.plot((ninetysixth, ninetysixth), (0, vertical), '--', linewidth=linewidth, color='C1')
plt.text(ninetysixth, vertical, "96th", ha='center', va='bottom', color='C1', size=textsize)
plt.plot((ninetyfifth, ninetyfifth), (0, vertical), '--', linewidth=linewidth, color='C1')
plt.text(ninetyfifth, vertical, "95th", ha='center', va='bottom', color='C1', size=textsize)
plt.yticks([])
plt.xticks([])
plt.axis("off")
plt.title("Example Baseline Density of Similarity Values", size=textsize)
plt.show()
```

```python
mem_reduced_distribution = scipy.stats.norm(0.8, 0.01)
yvalues_reduced = distances_distribution.pdf(xvalues) + mem_reduced_distribution.pdf(xvalues) / 10
ninetysixth_reduced = 0.77
```

```python
fig = plt.figure(figsize=(16, 3))
plt.plot(xvalues, yvalues_reduced)
plt.fill_between(xvalues, yvalues_reduced, color='lightblue')
plt.plot((ninetysixth_reduced, ninetysixth_reduced), (0, vertical), '--', linewidth=linewidth, color='C1')
plt.text(ninetysixth_reduced, vertical, "96th", ha='center', va='bottom', color='C1', size=textsize)
plt.plot((ninetyfifth, ninetyfifth), (0, vertical), '--', linewidth=linewidth, color='C1')
plt.text(ninetyfifth, vertical, "95th", ha='center', va='bottom', color='C1', size=textsize)
plt.yticks([])
plt.xlabel("Similarity ($\sigma$)", fontsize=20)
plt.xticks([])
plt.gca().spines['bottom'].set_visible(False)
plt.gca().spines['top'].set_visible(False)
plt.gca().spines['left'].set_visible(False)
plt.gca().spines['right'].set_visible(False)
plt.title("Example Density of Similarity Values After Memorization Reduction", size=textsize)
plt.show()
```
